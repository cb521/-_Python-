```python
import numpy as np;
import os;
import math
```

```python
# compute Mi(每个类的均值向量)and Ci(每个类的协方差矩阵)
# ω1：{(0 0)T, (2 0)T, (2 2)T, (0 2)T}
# ω2：{(4 4)T, (6 4)T, (6 6)T, (4 6)T}

'''----求m----'''
W=2  #类别总数

# w1=[[0,0],[2,0],[2,2],[0,2]]
# w2=[[4,4],[6,4],[6,6],[4,6]]
def Nav_Bayes(w1,w2,n1,n2,p_w1,p_w2,x1,x2):
    m1=[0,0]
    m2=[0,0]
    for i in range(n1):
        m1[0]=m1[0]+w1[i][0]
        m1[1]=m1[1]+w1[i][1]

    for i in range(n2):
        m2[0]=m2[0]+w2[i][0]
        m2[1]=m2[1]+w2[i][1]
    m1[0]=m1[0]/n1;
    m1[1]=m1[1]/n1;
    m2[0]=m2[0]/n2;
    m2[1]=m2[1]/n2;


    '''---求C---'''
    # 向量化
    w1=np.array(w1)
    w2=np.array(w2)

    m1=np.array(m1)[:,np.newaxis]
    m2=np.array(m2)[:,np.newaxis]


    C1=np.zeros((W,W))
    C2=np.zeros((W,W))
    for index,value in enumerate(w1):
        value=np.array(value)[:,np.newaxis]
        C1=C1+(value-m1).dot((value-m1).T)
    C1=C1/n1

    for index,value in enumerate(w2):
        value=np.array(value)[:,np.newaxis]
        C2=C2+(value-m2).dot((value-m2).T)
    C2=C2/n2

#     使用一类减二类的判别函数
    x=np.array([x1,x2])[:,np.newaxis]
    if (C1==C1).all():
        print("c1==c2")
        Const=-(0.5)*(m1.T).dot(np.linalg.inv(C1)).dot(m1)+(0.5)*(m2.T).dot(np.linalg.inv(C1)).dot(m2)+math.log(p_w1)-math.log(p_w2);
        L=((m1-m2).T).dot(np.linalg.inv(C1)).dot(x)+Const
        if L>=0:
            print("属于第一类");
        else:
            print("属于第二类");
    else:
        print("c1!=c2")
#         Const=-(0.5)*(m1.T).dot(np.linalg.inv(C1)).dot(m1)+(0.5)*(m2.T).dot(np.linalg.inv(C1)).dot(m2)+math.log(p_w1)-math.log(p_w2);
#         L=((m1-m2).T).dot(np.linalg.inv(C1)).dot(x)+Const
        d1=math.log(p_w1)-(0.5)*math.log(np.linalg.det(C1))-(0.5)*((x-m1).T).dot(np.linalg.inv(C1)).dot(x-m1)
        d2=math.log(p_w2)-(0.5)*math.log(np.linalg.det(C2))-(0.5)*((x-m2).T).dot(np.linalg.inv(C2)).dot(x-m2)
        if d1>d2:
            print("属于第一类");
        else:
            print("属于第二类");
```
